```python

import requests
import random

#url = "https://hls.ted.com/project_masters/8350/subtitles/it/full.vtt"
url = "https://hls.ted.com/project_masters/{}/subtitles/it/full.vtt" # 8350
```

```python
def sanitize(transcription):
    # remove first 2 lines
    transcription = transcription.split("\n")
    transcription = transcription[2:]
    transcription = "\n".join(transcription)

    # remove every empty line
    transcription = transcription.split("\n")
    transcription = [line for line in transcription if line != ""]
    transcription = "\n".join(transcription)

    # remove every line containing ' --> '
    transcription = transcription.split("\n")
    transcription = [line for line in transcription if " --> " not in line]
    transcription = "\n".join(transcription)

    # join every line with next line if it doesn't ends with a dot
    transcription = transcription.replace("\n", " ")
    transcription = transcription.split(". ")
    transcription = ".\n".join(transcription)

    return transcription
```

```python
for i in range(10000):
    r = requests.get(url.format(random.randint(1, 10000)))
    if r.status_code == 200:
        print("OK", i)
        with open("talks/{}.txt".format(i), "w+") as f:
            f.write(sanitize(r.text))
    else:
        print("ERROR")
```
