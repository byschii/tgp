```python
import requests
from bs4 import BeautifulSoup
import random
import time
```

```python
generate_year = lambda : random.randint(2019, 2023)
generate_month = lambda : random.randint(1, 12)
generate_day = lambda : random.randint(1, 28)

generale_link = lambda : f"https://www.open.online/{generate_year()}/{generate_month()}/{generate_day()}/"

def get_soup(url):
    time.sleep(1 + random.random()*2)
    resp = requests.get(url)
    if resp.status_code != 200:
        raise Exception(f"Error {resp.status_code} for url {url}, {resp.content}")
    soup = BeautifulSoup(resp.text, "html.parser")
    return soup

```

```python
# get
#  section#content
#  get all article.archieve
#  get figure
#  get a[href]

def get_urls_in_day(link):
    urls = []
    try:
        soup = get_soup(link)
        content = soup.find("div", {"class": "frontpage-open"})
        articles = content.find_all("article", {"class": "news"})

        for article in articles:
            a = article.find("figure").find("a")
            urls.append(a["href"])
    except Exception as e:
        print(e)

    return set(urls)
    
```

```python
urls = []
for i in range(124):
    res = get_urls_in_day(generale_link())
    urls += list(res)

print(len(urls))
```

```python

def scrape_article(url):
    time.sleep(1 + random.random()*3)
    content = ""
    title = ""
    try:
        resp = requests.get(url)
        assert resp.status_code == 200
        soup = BeautifulSoup(resp.text, "html.parser")

        # span#singleBody

        in_span = soup.find("span", {"class": "news__content"}) 
        in_div = soup.find("div", {"class": "news__content"})

        correct = in_span if in_span != None else in_div

        text_parts = correct.find_all("p", recursive=False)
        content = "\n".join([p.text for p in text_parts])
        title = url.split("/")[-1].replace("-", "_")
    except Exception as e:
        print(url, e)

    return title, content    

```

```python

for i, url in enumerate(urls):
    title, content = scrape_article(url)
    
    print(i, title, len(content))

    if len(content) == 0:
        continue
    
    with open(f"pages/{title}.txt", "w+") as f:
        f.write(content)



```
